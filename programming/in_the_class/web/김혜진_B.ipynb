- http://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=105

```python
from selenium import webdriver
import time
```

```python
driver = webdriver.Chrome()
```

```python
ls = []
```

```python
url = "http://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=105#&date=%2000:00:00&page="
ccs_path = ".section_body > ul > li > dl"
title_path = "dl > dt:not(.photo) > a"
for i in range(1, 11):
    time.sleep(3)
    driver.get(url + str(i))
    news = driver.find_elements_by_css_selector(ccs_path)
    for new in news:
        title = new.find_element_by_css_selector(title_path).text
        ls.append(title)
```
