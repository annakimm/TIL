설치된 폰트를 사용하는 방법은 크게 두가지

- rc parameter 설정으로 이후의 그림 전체에 적용
- 인수를 사용하여 개별 텍스트 관련 명령에만 적용

한글 문자열은 항상 유니코드를 사용해야 한다.

```python
# RC praameter 설정
%matplotlib inline
mpl.rc('font', family='Malgun Gothic')
# 그래프에서 마이너스 폰트 깨지는 문제 해결
mpl.rc('axes', unicode_minus=False)
```

```python
import numpy as np

data = np.random.randint(-100, 100, 50).cumsum()
data
```

```python
plt.plot(range(50), data, 'r')
mpl.rcParams['axes.unicode_minus'] = False
plt.title('시간별 가격 추이')
plt.ylabel('주식 가격')
plt.xlabel('시간(분)')
```

## 만약 개별적으로 폰트를 적용하고 싶을 때는 다음과 같이 폰트 패밀리, 색상, 크기를 정하여 플롯 명령의 fontdict 인수에 넣는다.

```python
font1 = {'family': 'HYGothic-Medium', 'color':  'black', 'size': 24}
font2 = {'family': 'HYGothic-Medium', 'color':  'darkred', 'weight': 'bold', 'size': 18}
font3 = {'family': 'HYGothic-Medium', 'color':  'blue', 'weight': 'light', 'size': 12}

x = np.linspace(0.0, 5.0, 100)
y = np.cos(2 * np.pi * x) * np.exp(-x)

plt.plot(x, y, 'k')
plt.title('한글 제목', fontdict=font1)
plt.xlabel('엑스 축', fontdict=font2)
plt.ylabel('와이 축', fontdict=font3)
plt.subplots_adjust()
plt.show()
```
