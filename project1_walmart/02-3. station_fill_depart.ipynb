```python
# interpolation 함수
wthr_interP = ["tmax", "tmin", "tavg", "dewpoint", "wetbulb", "sunset", "sunrise", "stnpressure", "sealevel"]
def fill_interP(obj, obj_fill):    
    a = obj[wthr_interP].interpolate(limit_direction="both", limit_area='inside')
    obj_fill.update(a)
```

```python
# median 함수
wthr_mean = ["depart", "snowfall", "preciptotal", "resultspeed", "resultdir", "avgspeed"]
def fill_median(obj, obj_fill):
    for i in wthr_mean:
        b = obj[i].replace(np.nan, obj[i].median())
        obj_fill.update(b)
```

```python
# interpolation과 median을 기존 함수에 합치기(station5 제외)
obj = [w_stn_1, w_stn_2, w_stn_3, w_stn_4, w_stn_5, w_stn_6, w_stn_7, w_stn_8, w_stn_9, w_stn_10, w_stn_11, w_stn_12,
       w_stn_13, w_stn_14, w_stn_15, w_stn_16, w_stn_17, w_stn_18, w_stn_19, w_stn_20]
obj_fill = [w_stn_1_fill, w_stn_2_fill, w_stn_3_fill, w_stn_4_fill, w_stn_5_fill, w_stn_6_fill, w_stn_7_fill, w_stn_8_fill, w_stn_9_fill,
           w_stn_10_fill, w_stn_11_fill, w_stn_12_fill,w_stn_13_fill, w_stn_14_fill, w_stn_15_fill, w_stn_16_fill, w_stn_17_fill,
           w_stn_18_fill, w_stn_19_fill, w_stn_20_fill]
for i in range(len(obj_fill)):
    fill_interP(obj[i], obj_fill[i])
    fill_median(obj[i], obj_fill[i])
```

```python
# heat, cool 다시 채우기
def fill_avg(obj_fill):
    obj_fill["tavg"] = round((obj_fill["tmin"] + obj_fill["tmax"])/2+0.5, 0)

def fill_heat_cool(obj_fill):  
    ls_c = []
    ls_h = []
    for temp in obj_fill["tavg"]:
        if temp > 65:
            c = temp - 65
            ls_c.append(c)
            ls_h.append(0)
        elif temp < 65:
            h = 65 - temp
            ls_c.append(0)
            ls_h.append(h)
        else:
            ls_c.append(0)
            ls_h.append(0)
    obj_fill["heat"] = ls_h
    obj_fill["cool"] = ls_c
```

```python
ls_hc = [w_stn_1_fill, w_stn_2_fill, w_stn_3_fill, w_stn_4_fill, w_stn_6_fill, w_stn_7_fill, w_stn_8_fill, w_stn_9_fill,
           w_stn_10_fill, w_stn_11_fill, w_stn_12_fill,w_stn_13_fill, w_stn_14_fill, w_stn_15_fill, w_stn_16_fill, w_stn_17_fill,
           w_stn_18_fill, w_stn_19_fill, w_stn_20_fill]
for i in range(len(ls_hc)):
    fill_heat_cool(ls_hc[i])
    fill_avg(ls_hc[i])
```

```python
# sunset, sunrise로 낮시간 계산
def fill_daytime(obj_fill):    
    df = obj_fill.copy()
    a = df["sunset"] - df["sunrise"]
    obj_fill["daytime"].update(a)
```

```python
ls_sun2 = [w_stn_2_fill, w_stn_3_fill, w_stn_4_fill, w_stn_5_fill, w_stn_6_fill, w_stn_10_fill, w_stn_11_fill, w_stn_14_fill, w_stn_15_fill, w_stn_18_fill, w_stn_19_fill]
for i in range(len(ls_sun2)):
    fill_daytime(ls_sun2[i])
```

```python
# 상대습도 계산
ls_rh = [w_stn_1_fill, w_stn_2_fill, w_stn_3_fill, w_stn_4_fill, w_stn_6_fill, w_stn_7_fill, w_stn_8_fill, w_stn_9_fill,
           w_stn_10_fill, w_stn_11_fill, w_stn_12_fill,w_stn_13_fill, w_stn_14_fill, w_stn_15_fill, w_stn_16_fill, w_stn_17_fill,
           w_stn_18_fill, w_stn_19_fill, w_stn_20_fill]
def fill_relative_humility(obj_fill):    
    b = 100*(np.exp((17.625*((obj_fill['dewpoint']-32)/1.8))/(243.04+((obj_fill['dewpoint']-32)/1.8)))/np.exp((17.625*((obj_fill['tavg']-32)/1.8))/(243.04+((obj_fill['tavg']-32)/1.8))))
    obj_fill["relative_humility"].update(b)
    
for i in range(len(ls_rh)):
    fill_relative_humility(ls_rh[i])
```

```python
# 체감온도 계산
ls_wc = [w_stn_1_fill, w_stn_2_fill, w_stn_3_fill, w_stn_4_fill, w_stn_6_fill, w_stn_7_fill, w_stn_8_fill, w_stn_9_fill,
           w_stn_10_fill, w_stn_11_fill, w_stn_12_fill,w_stn_13_fill, w_stn_14_fill, w_stn_15_fill, w_stn_16_fill, w_stn_17_fill,
           w_stn_18_fill, w_stn_19_fill, w_stn_20_fill]
def fill_windchill(obj_fill):    
    c = 35.74 + 0.6215*obj_fill["tavg"] - 35.75*(obj_fill["avgspeed"]**0.16) + 0.4275*obj_fill["tavg"]*(obj_fill["avgspeed"]**0.16)
    obj_fill["windchill"].update(c)
    
for i in range(len(ls_wc)):
    fill_windchill(ls_wc[i])
```
